### Cleaning data

```python
######## Duration & Rating ###########
#It can be seen clearly that an input error occured where the duration value was input for the rating
df.iloc[[5541,5794,5813]]

#Correcting the duration errors
df.loc[df.show_id=='s5542','duration']='74 min'
df.loc[df.show_id=='s5795','duration']='84 min'
df.loc[df.show_id=='s5814','duration']='66 min'

# Removing the error ratings value
df.loc[(df.show_id=='s5542')|(df.show_id=='s5795')|(df.show_id=='s5814'),'rating'] =None

# Viewing changes
df[df.rating.isin(['74 min','84 min','66 min'])| df.rating.isna()]

#print(f'There are {df.rating.unique().size-1} different ratings')
```

The content is grouped appropriately to allows for easy visualization.
United Stats has the highest count of movies and TV Shows. this is understanderbly because NEtflix is an American content with primiary viewership being the citizens. Also initial contents netflix used during its inception was for American consumption.
This India shows a large number of movies in second place, this is due to india being a country with largest number of customers. It becomes incumbant for Netflix to increase the number of indian movies so as to increase petronage.
However India is in 5th place for TV Shows with less than 100 shows whiles UK in second place with over 200 TV shows. This could be due to unavailability of Shows to add or Indian viewership prefer movies on netflix than series since netflix might prefer producing movies than TV Shows for the indian market.

```python
df_movies = pd.melt(df,id_vars=['type'],value_vars=['country'])
df_mov=df_movies.groupby(['type','value'],as_index=False).count().sort_values(['variable'],ascending=False)

#df_mov[df_mov['type']=='Movie'].head(10).plot('value','variable',kind='bar')
top_movies = df_mov[df_mov['type']=='Movie'].head(15)
top_series = df_mov[df_mov['type']=='TV Show'].head(15)
```

```python
fig,axes = plt.subplots(1,2,figsize=(16,6),sharex=False)
fig.suptitle('Top 15 Movies and Series')
sns.barplot(ax=axes[0],x="value", y="variable", data=top_movies,ci=None).set_xticklabels(axes[0].get_xticklabels(),rotation = 75);
sns.barplot(ax=axes[1],x="value", y="variable", data=top_series,ci=None).set_xticklabels(axes[1].get_xticklabels(),rotation = 75);
#labelling charts
axes[0].set_title('Top Movies');
axes[1].set_title('Top Series');
axes[0].set_xlabel('Country');
axes[1].set_xlabel('Country');
axes[0].set_ylabel('Quantity');
axes[1].set_ylabel('Quantity');
```

```python
fig,axes= plt.subplots(2,2,figsize=(12,6),sharey='row')
indian_movies=df[(df.type=='Movie') & (df.country=='India')]['date_added'].dt.year.value_counts(sort=False)
indian_movies.plot(kind='bar',ax=axes[0,0])
axes[0,0].set_title('Indian Movies ')
uk_movies=df[(df.type=='Movie') & (df.country=='United Kingdom')]['date_added'].dt.year.value_counts(sort=False)
uk_movies.plot(kind='bar',ax=axes[0,1])
axes[0,1].set_title('United Kingdom Movies ')

indian_series=df[(df.type=='TV Show') & (df.country=='India')]['date_added'].dt.year.value_counts(sort=False)
indian_series.plot(kind='bar',ax=axes[1,0])
axes[1,0].set_title('Indian TV Show ')
uk_series=df[(df.type=='TV Show') & (df.country=='United Kingdom')]['date_added'].dt.year.astype('Int64').value_counts(ascending=True)
uk_series.plot(kind='bar',ax=axes[1,1])
axes[1,1].set_title('United Kingdom TV Show ')

plt.suptitle('sSecond Highest Movie and TV Show')
plt.tight_layout();
```

```python

g=sns.catplot(
    data=content_duration[content_duration.release_year >2007], kind="bar",
    x="release_year",y='value_counts' ,hue="type",
    ci=None, palette="dark", alpha=.6, height=7,aspect=1.5
)
#g.despine(right=True)
g.set_axis_labels("", "NUmber of Media")
g.fig.suptitle("Content Distribution Over Years");
g.legend.set_title('Content Type')
#chartlabel(g)
```

MOvies and Series production is seen increasing  from 2008 and it peaked in 2017 with a decline from 2019. This could be attributed to the cost of making them  since movie making in the past decade requires huge sums of money. netflix being a suscriber based content provider cannot produce a large number of movies on its own but must also rely on other movie makers.
Now since 2016, there has been an increase in content providers like Disney+ etc which now prefer there platforms than host on Netflix for royaltes.
Finaly we can see a sturdy increase in the percentage of TV Shows until it supasses movies in 2021.
